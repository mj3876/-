```python
#17387 외적

def ccw(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    return (x2 - x1) * (y3 - y1) - (y2 - y1) * (x3 - x1)

def cross(l1, l2):
    p1=l1[0]
    p2=l1[1]
    p3=l2[0]
    p4=l2[1]
    p = ccw(p1,p2,p3) * ccw(p1,p2,p4)
    q = ccw(p3,p4,p1) * ccw(p3, p4, p2)

    # 일직선인 경우
    if p == 0 and q == 0:
        if l1[0] > l1[1]:
            l1 = (l1[1], l1[0])
        if l2[0] > l2[1]:
            l2 = (l2[1], l2[0])
        # 겹치는 부분 확인.,
        if l1[0] <= l2[1] and l2[0] <= l1[1]:
            return 1  # 일직선 위에서 겹침
        else:
            return 0  # 일직선 위지만 안 겹침
    else:
      if p <= 0 and q <= 0: #직각으로 만나는 경우 포함
        return 1
      else:
        return 0#..?


# 입력
x1, y1, x2, y2 = map(int, input().split())
L1 = [(x1, y1), (x2, y2)]

x3, y3, x4, y4 = map(int, input().split())
L2 = [(x3, y3), (x4, y4)]

print(cross(L1, L2))
```

```python
A = list(map(int, input().split()))
k = int(input())

def partition(A, p, r):
    pivot = A[r]
    L = p
    R = r - 1
    while L <= R:
        while L <= R and A[L] <= pivot:
            L += 1
        while L <= R and A[R] > pivot:
            R -= 1
        if L < R:
            A[L], A[R] = A[R], A[L]
    A[L], A[r] = A[r], A[L]
    return L

def select(A, p, r, k):
    if p == r:
        return A[p]
    q = partition(A, p, r)
    m = q - p + 1  # 왼쪽 부분의 원소 개수 (pivot 포함)
    if k == m:
        return A[q]
    elif k < m:
        return select(A, p, q - 1, k)
    else:
        return select(A, q + 1, r, k - m)

ans = select(A, 0, len(A) - 1, k)
print(ans)
```

```python
#32633 카드 뒤집기 게임
N = int(input())
A = list(map(int, input().split()))
if N == 1:
    print(1)
    exit()

end_run = [0] * N
start_run = [0] * N

end_run[0] = 1
for i in range(1, N):
    if A[i] == A[i - 1]:
        end_run[i] = end_run[i - 1] + 1
    else:
        end_run[i] = 1

start_run[N - 1] = 1
for i in range(N - 2, -1, -1):
    if A[i] == A[i + 1]:
        start_run[i] = start_run[i + 1] + 1
    else:
        start_run[i] = 1

overall_max_run = max(end_run)

prefix = [0] * (N + 1)#0,N-1개 배열의 2부터 N+1까지 최대길이 누적해서 저장
prefix[1]=1
current_length = 1
for i in range(2, N + 1):
    if A[i - 1] == A[i - 2]:
        current_length += 1
    else:
        current_length = 1
    prefix[i] = max(prefix[i - 1], current_length)

suffix = [0]*(N+2)
suffix[N]=1
current_length = 1
for i in range(N - 1, 0, -1):
    if A[i-1] == A[i]:
        current_length += 1
    else:
        current_length = 1
    suffix[i] = max(suffix[i + 1], current_length)

answer = overall_max_run
#여기까지 주어진 배열의 최대길이
#다음부터 최대길이가 되도록 뒤집음
for x in range(N + 1):
    prefix_part = prefix[x]
    if x < N:#오른쪽 배열
        suffix_part = suffix[x + 1]
        a = 0#이어붙인 길이
        if (1 - A[x - 1]) == A[x]:
            a = end_run[x-1] + start_run[x]
        b = max(prefix_part, suffix_part, a)#최대길이
    else:#마지막 원소->오른쪽 배열 없음
        suffix_part = 0
        a = 0
        b = max(prefix_part, suffix_part, a)

    if b > answer:
        answer = b

print(answer)
```

```python
#25682 체스판2
#이전 체스판문제와 다르게 cb,cw 변수에 저장X 배열에 저장한다
#시작값이 B고 이후에 BWBW가 반복되는것을 판단하는 배열=B배열,시작값이 W고 이후에 WBWB가 반복되는것을 판단하는 배열=W배열
#b에서 홀수인덱스-시작값이 B면->B배열에 1저장/W배열에 0저장
#                -시작값이 W면->B배열에 0저장/W배열에 1저장
#b에서 짝수인덱스-B면 시작값이 W라는 뜻->B배열에 0저장/W배열에 1저장
#                -W면 시작값이 B라는 뜻->B배열에 1저장/W배열에 0저장
N, M, K = map(int, input().split())
b = [input() for _ in range(N)]
B = [[0] * (M + 1) for _ in range(N + 1)]#나중에 원소 i와 i-1더하기 때문에 m,n보다 1행,1열씩 더 생성
W = [[0] * (M + 1) for _ in range(N + 1)]

for i in range(1, N + 1):
    for j in range(1, M + 1):#배열 B,W을 채운다
        if (i + j) % 2 == 0:
            if b[i-1][j-1]== 'B':
                B[i][j] = 1
                W[i][j] = 0
            else:
                B[i][j] = 0
                W[i][j] = 1
        else:
            if  b[i-1][j-1]== 'B':
                B[i][j] = 0
                W[i][j] = 1
            else:
                B[i][j] = 1
                W[i][j] = 0

        # 부분합계산
        B[i][j] += B[i-1][j] + B[i][j-1] - B[i-1][j-1]
        W[i][j] += W[i-1][j] + W[i][j-1] - W[i-1][j-1]

ans = 10**9
for i in range(1, N - K + 2):       # 1 ~ N-K+1
    for j in range(1, M - K + 2):   # 1 ~ M-K+1

        # sum:일치하는칸수=>최솟값=답
        Bsum = B[i+K-1][j+K-1] - B[i-1][j+K-1] - B[i+K-1][j-1] + B[i-1][j-1]

        Wsum = W[i+K-1][j+K-1] - W[i-1][j+K-1] - W[i+K-1][j-1] + W[i-1][j-1]

        ans = min(ans, Bsum, Wsum)

print(ans)
```

```python
#1106호텔고객수
#구하는것 최소비용
c, n = map(int, input().split())  # c: 목표 고객 수, n: 도시 개수
city = [tuple(map(int, input().split())) for _ in range(n)]  #(v,w)가치=비용,무게=고객

# 고객수는 최대 100명까지 늘어날 수 있음 → c 이상 100까지 DP 계산
MAX = 100+c #+c여야 빈리스트아니어서 에러 안남
dp = [10**9] * (MAX + 1)
dp[0] = 0  # 0명 모으는 데 드는 비용 = 0

for v,w in city:# 배낭문제에선 무게 마지노선이 있어서 1개 물건넣었을 때 j-w비교해야<->여긴 마지노선 없어서 c이상 고객을 받았을 때 최소비용만 저장
    for i in range(w, MAX + 1):
        dp[i] = min(dp[i], dp[i - w] + v)

# c 이상 중 최소 비용 찾기
answer = min(dp[c:MAX+1])#c부터, 물건 무게X
print(answer)
```
