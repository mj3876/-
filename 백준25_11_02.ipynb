```python
#1018
n,m=map(int,input().split())
b=[]
ans=[]
for i in range(n):
  b.append(input())
for col in range(n-7):#0부터 (n-7)-1까지 n=10이면 n=0,1,2까지
  for row in range(m-7):
    cb=0
    cw=0
    for i in range(col,col+8):
      for j in range(row,row+8):
        if (i+j)%2==0:
          if b[i][j]=='B':
            cb+=1
          else:
            cw+=1
        else:
          if b[i][j]=='B':
            cw+=1
          else:
            cb+=1
    ans.append(cb)
    ans.append(cw)
print(min(ans))
```

```python
#17357 표준편차->구간합 구하기
n = int(input())
a = list(map(int, input().split()))
arr = [1] * n

for i in range(n):
    for j in range(i):      # 0부터 i-1
        arr[i] = a[i] - arr[j]#아 맞다 이거 직전 원소끼리 비교해야되지..?a[i]-a[i-1]/차 구하는거 아니고 xi^2의 합-xi의 합^2

print(max(arr))#표준편차 공식~루트(제곱의 평균-평균의 제곱)
```
