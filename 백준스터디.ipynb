```python
#1965 상자넣
#문제:가장 긴 증가하는 부분 수열 LIS
#사용 알고리즘:다이나믹 프로그래밍 알고리즘
n = int(input())
a = list(map(int, input().split()))
dp=n*[1]
for i in range(n):
  for j in range(i):
    if a[i]>a[j]:#ex)i=4일때 j=i-1=3까지 반복->a[4]>a[3]
      dp[i]=max(dp[i],dp[j]+1)#dp[4]=1 dp[3]+1=2+1
print(max(dp))
```

```python
#12015
from bisect import bisect_left#파이썬 내장 이분탐색/binarysearch와 반환값은 동일,리스트에 넣을 값 인덱스 찾기vs리스트에 있는 값 인덱스 찾기 차이
n = int(input())
a = list(map(int, input().split()))
L = [] #L:길이값이 아닌 [10,20..] 이렇게 원소가 들어간 리스트
for x in a:#x:L에 넣을 원소
  p=bisect_left(L,x)#p:L에서 x가 들어갈 인덱스
  if p==len(L):
    L.append(x)
  else:#교체는 append 아닌 =
    L[p]=x
print(len(L))
```
